```python
import tensorflow as tf
import pandas as pd
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import preprocessing
```

```python
x = pd.read_csv('data_test_daily.csv')
x.head()
```

```python
y = x.pop('O3')
y.head()
```

```python
x = preprocessing.scale(x)
```

```python
model = tf.keras.models.load_model('ozone-final.model')
pred = model.predict(x)
```

```python
mse_train = mean_squared_error(y, pred)
rmse_train = sqrt(mean_squared_error(y, pred))
mae_train = mean_absolute_error(y, pred)
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y, pred)))
print("MSE: ",mse_train)
print("RMSE: ",rmse_train)
print("MAE: ",mae_train)
```
